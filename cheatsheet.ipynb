##Â Plots

````python
x = np.linspace(0,2,10)
y = np.sin(x*np.pi)

#Â Create figure with size in cm (each column in a two-column paper is
# around 9 cm wide)
x_cm = 8
y_cm = 5
fig, ax = plt.subplots(1,1,figsize=(x_cm/2.54, y_cm/2.54)) 
                     # 1,1 is the number of subplots columns and rows
                     # If more than one, then ax is a list. To get the
                     # axis of subplot (m,n), use ax[m][n].

# Plot
plt.plot(x, y, 
         color='tab:red', # 'tab:color' are nice flat colors
         alpha = 0.9, # transparency
         marker = 'o', markersize = 5, 
         markerfacecolor = 'black', 
         markeredgecolor = 'black',
         label = r'$N=10$ points')

# Axes limits
ax.set_xlim([-0.05,2.05])
plt.ylim([-1.05,1.05])

# Axes scales
plt.xscale('linear') # or 'log'
ax.set_yscale('linear') # or 'log'

# Axes labels
plt.xlabel(r'$x$-axis', fontsize=12)
ax.set_ylabel(r'$f_{\min}(x)$', fontsize=12)

# Axes ticks and tick labels
ax.set_xticks([0,0.8,1.6])
ax.set_yticks([-1,0,1])
ax.ticklabel_format(axis='x',style='plain') # or 'sci'
ax.ticklabel_format(axis='y',style='plain') # or 'sci'
ax.tick_params(labelsize=10)
                
#Â Legend
plt.legend()

# Show or save
save = False
if save:
    filename = 'figs_directory/figname_xcm%d_ycm%s_xin%.2f.pdf'%(x_cm,
                                                                 y_cm,
                                                                 1.234)
    plt.savefig(filename,dpi=300,bbox_inches='tight')
else:
    plt.show()
````

#### Two different vertical axes:

````python
x = np.linspace(0,2,20)
y = np.sin(x*np.pi)

fig, ax = plt.subplots()

# Plot
plt.plot(x, y, color='black')

# Second axis
ax2 = ax.twinx()
ax2.scatter(x, -y, marker='D', color='tab:red')
ax2.set_ylabel(r'Second $y$-axis', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.spines['right'].set_color('tab:red')
ax2.tick_params(axis='y', which='both', colors='tab:red')

plt.show()
````

#### Colors from a colormap:

````python
x = np.linspace(0,2,50)
y = [x]*4
for i in range(0,len(y)):
    y[i] = np.sin(x*np.pi+i*np.pi/4)

# Get colors from colormap
cmap = plt.cm.get_cmap('inferno')
colors = [cmap(0),cmap(0.5),cmap(0.75),cmap(0.875)]

# Plot\mathrm{u}
fig, ax = plt.subplots()

for i in range(0,len(y)):
    plt.plot(x, y[i], color=colors[i])

plt.show()
````

#### Text in LaTeX expressions:

LaTeX command `\text` does not work in `matplotlib` because it requires the amsmath package (not included in `mathtext`, the math rendering engine of `matplotlib`). Instead, use `\mathrm`.

````python
x = np.linspace(0,2,50)
y = np.sin(x*np.pi)

plt.plot(x,y)

plt.xlabel(r'Equation: $A_\mathrm{opt}-\varepsilon^\mathrm{(test)}$',
           fontsize=20)

plt.show()
````
